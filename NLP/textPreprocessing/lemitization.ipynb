```python
words = [
        "program", "programs", "programmer", "programming", "programmers",
        "playing", "played", 
        "friends", "friendship",
        "ponies", "universal", "university", "alumni", "alu"
    ]
```

pos -Nounn -n 
 verb- V
 adjective-a
 adverb-r

```python
from nltk.stem import WordNetLemmatizer

lemmatizer=WordNetLemmatizer()
```

```python
for word in words:
    print(word+"---->>"+lemmatizer.lemmatize(word, pos='v'))
```
