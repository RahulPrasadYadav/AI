```python
words = [
        "program", "programs", "programmer", "programming", "programmers",
        "playing", "played", 
        "friends", "friendship",
        "ponies", "universal", "university", "alumni", "alu"
    ]
```

```python
from nltk.stem import PorterStemmer

porter_stem=PorterStemmer()

for word in words:
      print(word+"----->"+porter_stem.stem(word))
```

##  RegexepStemmer

```python
from nltk.stem import RegexpStemmer

reg_stem=RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stem.stem('eating')
```
