```python
corpus="""   hello  wellcome, to rahul prasad  tutorials.

pelase do the watch the entire course! to become expert  in nlp.

"""
```

```python
print(corpus)
```

```python
from nltk.tokenize import sent_tokenize


docoment=sent_tokenize(corpus)
```

```python
print(docoment)
```

```python
print(type(docoment))
```

```python
for sen in docoment:
    print(sen)
```

## sentence --> word 
## paragraph --> word 

```python
from nltk.tokenize import word_tokenize

word_tokenize(corpus)
```

```python
for sen in docoment:
    print(word_tokenize(sen))
```
